```python
# OpenCV-Python：图像的几何变换（平移、旋转、仿射变换等）
```

```python
import cv2
import numpy as np
```

```python
# 平移变换
img = cv2.imread('images/iron_man.jpg',1)
rows,cols,channel = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img, M, (cols,rows))

cv2.imshow('img', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 旋转变换
img = cv2.imread('images/iron_man.jpg',1)
rows,cols,channel = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/3), 90, 0.8)
dst = cv2.warpAffine(img, M, (cols, rows))

cv2.imshow('img', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 仿射变换---二维到二维
from matplotlib import pyplot as plt

img = cv2.imread('images/iron_man.jpg',1)
rows,cols,channel = img.shape

pts1 = np.float32([[50,50], [200,50], [50,200]])
pts2 = np.float32([[10,100], [200,50],[100,250]])

M = cv2.getAffineTransform(pts1, pts2)

dst = cv2.warpAffine(img, M, (cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
# 透射变换
img = cv2.imread('images/iron_man.jpg',1)
rows,cols,channel = img.shape

pts1 = np.float32([[56,55], [368,52], [28,387], [389,390]])
pts2 = np.float32([[0,1], [300,0],[0,300], [300,300]])

M = cv2.getPerspectiveTransform(pts1, pts2)

dst = cv2.warpPerspective(img, M, (300,300))

plt.subplot(121),plt.imshow(img),plt.title('Imput')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
